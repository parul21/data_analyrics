```python
import nltk
import pandas as pd
dataset = pd.read_csv('tweets.csv','r',error_bad_lines=False)
```

```python
pat=r'''#\w+'''
tokenizer = nltk.regexp_tokenize(str(dataset),pat)
tokenizer
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
tokenizer=list(dict.fromkeys(tokenizer))
tokenizer
```

```python
stemtoken=[]
for token in tokenizer:
    print(ps.stem(token))
    stemtoken.append(ps.stem(token))

```

```python
from nltk import ngrams
n=3
ngrm=[]
res = ngrams(str(dataset).split(),n) 
for s in res:
    ngrm.append(s)
ngrm
```
