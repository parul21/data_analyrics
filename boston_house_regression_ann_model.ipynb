# **boston house regression ann model**

```
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.datasets import load_boston 
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
```

```
bhouse=load_boston()
```

```
print(bhouse.feature_names)
print(bhouse.DESCR)
```

```
input_data=bhouse.data
target_class=bhouse.target 
```

```
scaler = StandardScaler()
input_scaled = scaler.fit_transform(input_data)
print(input_scaled)
```

```
x_train,x_test,y_train,y_test = train_test_split(input_scaled,target_class,test_size=0.3)
print(x_train)
```

```
model = Sequential()
h1=model.add(Dense(50, activation='relu'))
h2=model.add(Dense(100, activation='relu'))
h3=model.add(Dense(50, activation='relu'))
o=model.add(Dense(1))
```

```
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['accuracy'])
```

```
model.fit(x_train,y_train,batch_size=10,epochs=1000,shuffle=True,verbose=2)
```

```
model.summary()
```

```
score = model.evaluate(x_test,y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
